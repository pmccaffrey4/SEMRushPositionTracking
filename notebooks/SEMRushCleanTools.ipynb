```python
import pandas as pd


filename = "/home/patrick/PycharmProjects/SEMRushPositionTracking/smartmouth_1108_1219_2020_positions.xlsx"
df = pd.read_excel(filename, skiprows=7, sheet_name="Data")


```

```python
df.head()
```

```python
rank_cols = [col for col in df.columns if not any(x in col for x in ['type', 'landing', 'Tags', 'Search',
                                                                     'CPC', 'difference'])]
```

```python

rank_cols[:10]
```

```python

clean_df = df.copy()[rank_cols]
```

```python
clean_df.head()
```

```python
for col in clean_df.columns:
    clean_df.rename(columns={col:col.strip('*.smartmouth.com/*_')}, inplace=True)
    
    
    # df.rename(columns={col:col.upper().replace(" ","_")},inplace=True)
```

```python
df_melt = pd.melt(clean_df, id_vars=['Keyword'], value_vars=list(clean_df.columns[1:]), var_name='date', value_name='position')
```

```python
df_melt['date'] = pd.to_datetime(df_melt['date'])
```

```python
df_melt.head()
```

```python
df_melt.to_csv('smartmouth.csv')
```
